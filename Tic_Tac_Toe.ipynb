```python
import copy
```

```python
class TicTacToe:
    def __init__(self):
        self.board = [[None for _ in range(3)] for _ in range(3)]
        self.start = 'x'
        self.winner = None
        self.moves_left = 9

    def new_game(self):
        self.board = [[None for _ in range(3)] for _ in range(3)]
        self.start = 'x'
        self.winner = None
        self.moves_left = 9

        mode = int(input("Enter 1 if you want to play with AI 2 otherwise: "))
        if(mode == 1):
            self.play_with_AI()
        else:
            self.play()

    def print_board(self):
        print("  0 1 2")
        print("  ------")
        for i in range(3):
            print(f"{i}|", end = "")
            for j in range(3):
                print(" " if self.board[i][j] is None else self.board[i][j], end = " ")
            print("|")

        print("  ------")



    def check_winner(self, new_board):
        #check for rows #
        for i in range(3):
            if(new_board[i][0] == new_board[i][1] == new_board[i][2] and new_board[i][0] is not None):
                return new_board[i][0]

        # check for columns
        for i in range(3):
            if(new_board[0][i] == new_board[1][i] == new_board[2][i] and new_board[0][i] is not None):
                return new_board[0][i]

        # check diagonals
        if(new_board[0][0] == new_board[1][1] == new_board[2][2] and new_board[0][0] is not None):
            return new_board[0][0]

        if(new_board[0][2] == new_board[1][1] == new_board[2][0] and new_board[0][2] is not None):
            return new_board[0][2]

        return None


    def play_with_AI(self):
        while(self.winner is None):
            self.print_board()
            valid_move = False
            while(valid_move == False):
                if(self.start == 'o'):
                    x = int(input("Enter your desired row to place sticker: "))
                    y = int(input("Enter your desired column to place sticke: "))
                    if(x>=0 and y>=0 and x<3 and y<3 and self.board[x][y] is None):
                        self.board[x][y] = self.start
                        self.moves_left -= 1
                        self.winner = self.check_winner(self.board)
                        valid_move = True
                    else:
                        print("Invalid Move detected!")
                else:
                    score, m = self.minimax(self.board, self.start)
                    if m is not None:
                        self.board[m[0]][m[1]] = self.start
                        self.moves_left -= 1
                        self.winner = self.check_winner(self.board)
                        valid_move = True
                    else:
                        print("No valid moves left")
            if(self.winner is not None):
                self.print_board()
                print(f"Winner Found {self.start}")
                break
            self.start = 'x' if self.start == 'o' else 'o'
            # self.print_board()
            if(self.winner is None and self.moves_left == 0):
                print("Game Draw!")
                break

    def play(self):
        while(self.winner is None):
            self.print_board()
            valid_move = False
            while(valid_move == False):
                x = int(input("Enter your desired row to place sticker: "))
                y = int(input("Enter your desired column to place sticke: "))
                if(x>=0 and y>=0 and x<3 and y<3 and self.board[x][y] is None):
                    self.board[x][y] = self.start
                    self.moves_left -= 1
                    self.winner = self.check_winner(self.board)
                    valid_move = True
                else:
                    print("Invalid Move detected!")
            if(self.winner is not None):
                self.print_board()
                print(f"Winner Found {self.start}")
                break
            self.start = 'x' if self.start == 'o' else 'o'
            # self.print_board()
            if(self.winner is None and self.moves_left == 0):
                print("Game Draw!")
                break

    def minimax(self, board, curr_player, alpha = -10000, beta=10000, depth = 0):
        winner = self.check_winner(board)
        if(winner is not None):
            if(winner == 'x'):
                return 10, None
            elif(winner == 'o'):
                return -10, None

        is_full = True
        for i in range(3):
            for j in range(3):
                if(board[i][j] is None):
                    is_full = False

        if(is_full):
            return 0, None

        best_score = -1000 if curr_player == 'x' else 1000
        best_move = None
        for i in range(3):
            for j in range(3):
                if(board[i][j] is None):
                    new_board = copy.deepcopy(board)
                    new_board[i][j] = curr_player
                    score,_ = self.minimax(new_board, 'x' if curr_player == 'o' else 'o',alpha, beta, depth+1)

                    if curr_player == 'x':
                        if score>best_score:
                            best_score = score
                            best_move = (i,j)
                            alpa = max(alpha, best_score)
                    else:
                        if score<best_score:
                            best_score = score
                            best_move = (i,j)
                            beta = min(beta, best_score)

                    if(beta<=alpha):
                        break

        return best_score, best_move
```
